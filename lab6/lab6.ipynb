```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def convert_to_binary_image(image_path):
    image = Image.open(image_path).convert('1')  # Convert to binary
    return np.array(image)

def embed_watermark(input_image_path, watermark_binary, bit_plane):
    image = Image.open(input_image_path)
    image_array = np.array(image)
    watermark_rows, watermark_cols = watermark_binary.shape
    img_rows, img_cols, _ = image_array.shape

    for i in range(img_rows):
        for j in range(img_cols):
            watermark_pixel = watermark_binary[i % watermark_rows, j % watermark_cols]
            bit = 1 if watermark_pixel else 0

            image_array[i, j, 2] = (image_array[i, j, 2] & ~(1 << (bit_plane - 1))) | (bit << (bit_plane - 1))

    return Image.fromarray(image_array)

def text_to_bits(text):
    return ''.join(format(ord(char), '08b') for char in text)

def embed_text(input_image, text):
    image_array = np.array(input_image)
    binary_text = text_to_bits(text)
    data_index = 0
    text_length = len(binary_text)

    for i in range(image_array.shape[0]):
        for j in range(image_array.shape[1]):
            if data_index < text_length:
                image_array[i, j, 2] = (image_array[i, j, 2] & 0xFE) | int(binary_text[data_index])
                data_index += 1

    return Image.fromarray(image_array)

def remove_watermark(input_image_path, bit_plane):
    image = Image.open(input_image_path)
    image_array = np.array(image)
    rows, cols, _ = image_array.shape

    for i in range(rows):
        for j in range(cols):
            image_array[i, j, 2] = image_array[i, j, 2] & ~(1 << (bit_plane - 1))

    return Image.fromarray(image_array)

def extract_text(image, text_length):
    image_array = np.array(image)
    binary_text = ""
    
    for i in range(image_array.shape[0]):
        for j in range(image_array.shape[1]):
            binary_text += str(image_array[i, j, 2] & 1)
            if len(binary_text) >= text_length * 8:
                break
        if len(binary_text) >= text_length * 8:
            break

    text = ''
    for i in range(0, len(binary_text), 8):
        byte = binary_text[i:i+8]
        text += chr(int(byte, 2))

    return text
```

```python
watermark_binary = convert_to_binary_image('WM.png')
plt.imshow(watermark_binary, cmap='gray')
plt.title("Binary Watermark")
plt.axis('off')
plt.show()
```

```python
watermarked_image = embed_watermark('InputImage.png', watermark_binary, 8)
plt.imshow(watermarked_image)
plt.title("Watermarked Image")
plt.axis('off')
plt.show()
```

```python
embedded_text_image = embed_text(watermarked_image, "Hello world!")
plt.imshow(embedded_text_image)
plt.title("Image with Embedded Text")
plt.axis('off')
plt.show()
```

```python
unwatermarked_image = remove_watermark('InputImage.png', 6)
plt.imshow(unwatermarked_image)
plt.title("Unwatermarked Image")
plt.axis('off')
plt.show()
```

```python
extracted_text = extract_text(embedded_text_image, len("Hello world!"))
print("Extracted text:", extracted_text)
```
