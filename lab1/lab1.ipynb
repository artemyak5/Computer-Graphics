<p style="font-size:15px;">Задати вершини довільного опуклого 8-кутника на площині.</p>

```python
osnova = np.array([[3, 4], 
                   [3, 5], 
                   [4, 3], 
                   [4, 6], 
                   [5, 3], 
                   [5, 6], 
                   [6, 4], 
                   [6, 5]])
```

<p style="font-size:15px;">Візуалізуємо</p>

```python
def plot_points(points):
    x = points[:, 0]
    y = points[:, 1]

    plt.scatter(x, y, color='green')

    plt.axhline(y=0, color='gray', linestyle='--')
    plt.axvline(x=0, color='gray', linestyle='--')

    plt.xlabel('X')
    plt.ylabel('Y')

    plt.xlim(-8, 8)
    plt.ylim(-8, 8)

    plt.grid(True)
    
plot_points(osnova)
```

 <p style="font-size:15px;">Зменшити його в 1,5 рази.</p>

```python
scaled_points = osnova / 1.5
plot_points(scaled_points)
```

<hr>
<center style="font-size:18px;"><b> Завдання №3 </b></center>
<hr>

<p style="font-size:15px;">Отриманий результат з попереднього пункту симетрично відобразити відносно початку координат.</p>

```python
transformation_matrix = np.array([[-1, 0, 0],
                                  [0, -1, 0],
                                  [0, 0, 1]])  
symmetric_points = np.dot(scaled_points, transformation_matrix[:, :-1])
plot_points(symmetric_points)
```

```python
def pryama(x):    
    plot_points(x)
    x1, y1 = [0, -2]
    x2, y2 = [2, 0]

    x_values = np.linspace(x1 - 10, x2 + 10, 100)

    a = (y2 - y1) / (x2 - x1)
    b = y1 - a * x1

    y_values = a * x_values + b

    plt.plot(x_values, y_values, color='black')

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.show()
    
pryama(scaled_points)
```
