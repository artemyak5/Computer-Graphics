```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<hr>
<center style="font-size:18px;"><b> Завдання №2 </b></center>
<hr>

<p style="font-size:15px;">Задати вершини довільного опуклого 8-кутника на площині.</p>

```python
octagon = np.array([[3, 4],
                   [3, 5],
                   [4, 6],
                   [6, 6],
                   [6, 5],
                   [6, 4],
                   [5, 3],
                   [4, 3]])
```

<p style="font-size:15px;">Візуалізуємо</p>

```python
def plot_points(points):
    x = points[:, 0]
    y = points[:, 1]

    plt.scatter(x, y, color='green')

    plt.axhline(y=0, color='gray', linestyle='--')
    plt.axvline(x=0, color='gray', linestyle='--')

    plt.xlabel('X')
    plt.ylabel('Y')

    plt.xlim(-8, 8)
    plt.ylim(-8, 8)

    plt.grid(True)
    
plot_points(octagon)
```

 <p style="font-size:15px;">Зменшити його в 1,5 рази.</p>

```python
scaled_points = octagon / 1.5
scaled_points = np.hstack((scaled_points, np.ones((scaled_points.shape[0], 1))))
plot_points(scaled_points)
```

<hr>
<center style="font-size:18px;"><b> Завдання №3 </b></center>
<hr>

<p style="font-size:15px;">Отриманий результат з попереднього пункту симетрично відобразити відносно початку координат.</p>

```python
transformation_matrix = np.array([[-1, 0, 0],
                                  [0, -1, 0],
                                  [0, 0, 1]])  
symmetric_points = np.dot(scaled_points, transformation_matrix[:, :-1])
plot_points(symmetric_points)
```

<hr>
<center style="font-size:18px;"><b> Завдання №4 </b></center>
<hr>

<p style="font-size:15px;">Задати похилу пряму двома точками на площині.</p>

```python
point1 = np.array([0, -2, 1])
point2 = np.array([6, 4, 1])

k = (point2[1] - point1[1]) / (point2[0] - point1[0])
b = point1[1] - k*point1[0]

plot_points(scaled_points)

x_values = [point1[0], point2[0]]
y_values = [point1[1], point2[1]]

plt.plot(x_values, y_values, color='black')

plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
phi = np.arctan(k)

# rotateZ = np.array([
#     [np.cos(phi), -np.sin(phi), 0],
#     [np.sin(phi), np.cos(phi), 0],
#     [0, 0, 1]
# ])

# rotateY = np.array([
#     [np.cos(phi), 0, np.sin(phi)],
#     [0, 1, 0],
#     [-np.sin(phi), 0, np.cos(phi)]
# ])

# rotateX = np.array([
#     [1, 0, 0],
#     [0, np.cos(phi), -np.sin(phi)],
#     [0, np.sin(phi), np.cos(phi)]
# ])

#translationMatrix
M1 = np.array([[1, 0, 0],
               [0, 1, 0],
               [0, -b, 1]])


#rotationMatrix
M2 = np.array([[np.cos(phi), np.sin(phi), 0],
               [-np.sin(phi), np.cos(phi), 0],
               [0, 0, 1]])
#reflectXMatrix
M3 = np.array([[1, 0, 0],
               [0, -1, 0],
               [0, 0, 1]])

#rotationBackMatrix
M4 = np.array([[np.cos(-phi), np.sin(-phi), 0],
               [-np.sin(-phi), np.cos(-phi), 0],
               [0, 0, 1]])

#translationBackMatrix
M5 = np.array([[ 1, 0, 0],
               [0, 1, 0],
               [0, b, 1]])

reflexByLine = M1.dot(M4).dot(M3).dot(M2).dot(M5)
operation = scaled_points.dot(reflexByLine)
tree = np.vstack((scaled_points, operation))

plot_points(tree)
plt.plot(x_values, y_values, color='black')
```
