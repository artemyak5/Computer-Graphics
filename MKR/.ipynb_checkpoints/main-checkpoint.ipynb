```python
import numpy as np
import matplotlib.pyplot as plt
import copy
import math



pram = 5

# Створення масивів координат
center = pram // 2

# Початкова матриця
matrix = np.zeros((pram, pram), dtype=int)

for i in range(pram):
    matrix[center][i] = 1 
    matrix[i][center] = 1 

# Створення матриці S та обчислення фрактальної розмірності
S = 1 - matrix
size = matrix.shape[0]
non_zero_elements = np.sum(matrix)
fractal_dimension = math.log(non_zero_elements) / math.log(size)
plt.title(f'{np.floor(fractal_dimension)} < {fractal_dimension:.2f} < {np.ceil(fractal_dimension)}')
plt.imshow(S, cmap='gray')
plt.show()
```

```python
# Рекурсивне створення матриці
for _ in range(2):
    matrix_list = matrix.tolist()
    matrix_copy = copy.deepcopy(matrix_list)
    non_zero_elements = 0
    for i in range(len(matrix_list)):
        for j in range(len(matrix_list[0])):
            if matrix_list[i][j] == 1:
                non_zero_elements += 1
                matrix_list[i][j] = np.array(matrix_copy)
            else:
                matrix_list[i][j] = np.zeros_like(matrix_copy)
    matrix = np.block(matrix_list)
    S = 1 - matrix
    size = matrix.shape[0]
    fractal_dimension = math.log(non_zero_elements) / math.log(size)
    plt.title(f'{np.floor(fractal_dimension)} < {fractal_dimension:.2f} < {np.ceil(fractal_dimension)}')
    plt.imshow(S, cmap='gray')
    plt.show()
```
