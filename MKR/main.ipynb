```python
import numpy as np
import matplotlib.pyplot as plt
import copy
import math

base_size = 5

excluded_columns = [1, 3]
excluded_rows = [0, 2, 4]

columns = np.linspace(0, base_size-1, base_size, dtype=int)
rows = np.linspace(0, base_size-1, base_size, dtype=int)
columns = np.delete(columns, excluded_columns)
rows = np.delete(rows, excluded_rows)

initial_matrix = np.zeros((base_size, base_size), dtype=int)
non_zero_count = 0
for col in columns:
    for row in rows:
        initial_matrix[row][col] = 1
        non_zero_count += 1

complement_matrix = 1 - initial_matrix
matrix_size = initial_matrix.shape[0]
initial_dimension = math.log(non_zero_count) / math.log(matrix_size)

plt.title(f'{np.floor(initial_dimension)} < {initial_dimension:.2f} < {np.ceil(initial_dimension)}')
plt.imshow(complement_matrix, cmap='gray')
plt.show()
```

```python
for _ in range(2):
    matrix_list = initial_matrix.tolist()
    matrix_copy = copy.deepcopy(matrix_list)
    non_zero_count = 0
    for i in range(len(matrix_list)):
        for j in range(len(matrix_list[0])):
            if matrix_list[i][j] == 1:
                non_zero_count += 1
                matrix_list[i][j] = np.array(matrix_copy)
            else:
                matrix_list[i][j] = np.zeros_like(matrix_copy)
    initial_matrix = np.block(matrix_list)
    complement_matrix = 1 - initial_matrix
    matrix_size = initial_matrix.shape[0]
    fractal_dimension = math.log(non_zero_count) / math.log(matrix_size)
    plt.title(f'{np.floor(fractal_dimension)} < {fractal_dimension:.2f} < {np.ceil(fractal_dimension)}')
    plt.imshow(complement_matrix, cmap='gray')
    plt.show()
```
