<center style="font-size: 16px"><b>Завдання 3</b></center>
<p style="font-size: 15px">Обрати 8 будь-яких із запропонованих фракталів та реалізувати їх (написати відповідну програму). Врахувати те, що для побудови фрактальної кривої використовується змінна F та символи «+» і «–». Інші змінні у даних фракталах використовуються для побудови співвідношень, але не для самої візуалізації.</p>

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_fractal(axiom, rules, angle, iterations, L=1.0, starting_angle=0, line_width=1.0):
    state = axiom
    stack = []
    x, y = 0.0, 0.0
    current_angle = starting_angle

    plt.figure(figsize=(10, 10))
    plt.axis('equal')
    plt.axis('off')
    
    for _ in range(iterations):
        next_state = ""
        for character in state:
            next_state += rules.get(character, character)
        state = next_state

    for character in state:
        if character == 'F':
            next_x = x + L * np.cos(np.radians(current_angle))
            next_y = y + L * np.sin(np.radians(current_angle))
            plt.plot([x, next_x], [y, next_y], color="black", linewidth=line_width)
            x, y = next_x, next_y
        elif character == '+':
            current_angle += angle
        elif character == '-':
            current_angle -= angle
        elif character == '[':
            stack.append((x, y, current_angle))
        elif character == ']':
            x, y, current_angle = stack.pop()

    plt.show()
```

<hr>
<center style="font-size: 16px"><b>3.1 Хрестоподібний фрактал</b></center>

```python
axiom = "F+XF+F+XF"
rules = {
    "X": "XF-F+F-XF+F+XF-F+F-X"
}
angle = 90

draw_fractal(axiom, rules, angle, iterations=3, line_width=0.5)
```

<p style="font-size: 15px">Модифікувати програму з пунктів 3 і 4 даних завдань для побудови
фракталів з кутом θmin ≤ θ ≤ θmax. Зробити відображення у вигляді
плавного переходу для вказаного діапазону θ ∈ [θmin; θmax].</p>

```python
from matplotlib.animation import FuncAnimation, PillowWriter

#
def update_frame(i, angles, iterations, line_width=1.0):
    # 
    plt.clf()
    plt.axis('equal')
    plt.axis('off')

    axiom = "F"
    rules = {
    "F": "-F++F-"
    }
    state = axiom
    stack = []
    x, y = 0.0, 0.0
    current_angle = 0
    angle = angles[i]

    for _ in range(iterations):
        next_state = ""
        for character in state:
            next_state += rules.get(character, character)
        state = next_state

    for character in state:
        if character == 'F':
            next_x = x + np.cos(np.radians(current_angle))
            next_y = y + np.sin(np.radians(current_angle))
            plt.plot([x, next_x], [y, next_y], color="black", linewidth=line_width)
            x, y = next_x, next_y
        elif character == '+':
            current_angle += angle
        elif character == '-':
            current_angle -= angle


fig, ax = plt.subplots()
angles = np.linspace(90, 45, 20) 
iterations = 15  

ani = FuncAnimation(fig, update_frame, frames=len(angles), fargs=(angles, iterations), interval=400)

writer = PillowWriter(fps=10)  # Встановлення швидкості кадрів на секунду
ani.save("dragon_curve_animation400.gif", writer=writer)

plt.show()
```
