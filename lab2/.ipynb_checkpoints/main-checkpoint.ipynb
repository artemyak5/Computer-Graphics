```python
vertices = np.array([[1, 1, 1],
                     [2, 1, 1],
                     [2, 2, 1],
                     [1, 2, 1],
                     [1, 1, 2],
                     [2, 1, 2],
                     [2, 2, 2],
                     [1, 2, 2]])

edges = [[0, 1], [1, 2], [2, 3], [3, 0],
         [4, 5], [5, 6], [6, 7], [7, 4],
         [0, 4], [1, 5], [2, 6], [3, 7]]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Відображаємо вершини
# ax.scatter3D(vertices[:, 0], vertices[:, 1], vertices[:, 2])

for edge in edges:
    ax.plot3D(*zip(*vertices[edge]), color='b')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
def draw(vertices):
    
    edges = [[0, 1], [1, 2], [2, 3], [3, 0],
         [4, 5], [5, 6], [6, 7], [7, 4],
         [0, 4], [1, 5], [2, 6], [3, 7]]
    
    fig = plt.figure()
    for edge in edges:
        edge_vertices = vertices[edge]
        ax.plot3D(*zip(*edge_vertices), color='b')

# Відображаємо вершини
# ax.scatter3D(vertices[:, 0], vertices[:, 1], vertices[:, 2])

    for edge in edges:
        ax.plot3D(*zip(*vertices[edges]), color='b')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()    
```

```python
def scaled(scaling_factor):
    scaling_matrix = np.array([[scaling_factor, 0, 0],
                               [0, scaling_factor, 0],
                               [0, 0, scaling_factor]])

    scaled_vertices = vertices.dot(scaling_matrix)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Відображаємо з'єднання вершин для звичайного куба
    for edge in edges:
        ax.plot3D(*zip(*vertices[edge]), color='b')

    # Відображаємо з'єднання вершин для масштабованого куба
    for edge in edges:
        ax.plot3D(*zip(*scaled_vertices[edge]), color='r')
        
    # Позначаємо осі координат
    ax.plot3D([-4, 4], [0, 0], [0, 0], color='black')  # Ось x (від 0 до 1.5)
    ax.plot3D([0, 0], [-4, 4], [0, 0], color='black')  # Ось y (від 0 до 1.5)
    ax.plot3D([0, 0], [0, 0], [-4, 4], color='black')  # Ось z (від 0 до 1.5)
    
    


    # Налаштування відображення
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()
    
scaled(2)
scaled(0.5)
```

```python
def trans():
    M_z = np.array([[1, 0, 0],
              [0, 1, 0],
              [0, 0, -1]])

    vertices_zz = vertices.dot(M_z)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Відображаємо з'єднання вершин для звичайного куба
    for edge in edges:
        ax.plot3D(*zip(*vertices[edge]), color='b')

    # Відображаємо з'єднання вершин для масштабованого куба
    for edge in edges:
        ax.plot3D(*zip(*vertices_zz[edge]), color='r')
        
    # Позначаємо осі координат
    ax.plot3D([-4, 4], [0, 0], [0, 0], color='black')  # Ось x (від 0 до 1.5)
    ax.plot3D([0, 0], [-4, 4], [0, 0], color='black')  # Ось y (від 0 до 1.5)
    ax.plot3D([0, 0], [0, 0], [-4, 4], color='black')  # Ось z (від 0 до 1.5)
    
    


    # Налаштування відображення
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()
    
trans()
```
