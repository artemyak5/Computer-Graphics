<p>3. Програмно реалізувати усі фрактали з підрозділів 1.4 та 1.5 даної лабораторної роботи.</p>

```python
def generate_fractal(iterations: int, params: list) -> tuple:
    
    x, y = [0], [0]
    if 'a' in params[0]:
        probabilities = [param.get('p', 1 / len(params)) for param in params]
        for _ in range(iterations):
            param = np.random.choice(params, p=probabilities)
            x1 = param['a'] * x[-1] + param['b'] * y[-1] + param['e']
            y1 = param['c'] * x[-1] + param['d'] * y[-1] + param['f']
            x.append(x1)
            y.append(y1)
    else:
        probabilities = [param.get('p', 1 / len(params)) for param in params]
        for _ in range(iterations):
            param = np.random.choice(params, p=probabilities)
            x1 = param['r'] * np.cos(param['theta']) * x[-1] - param['s'] * np.sin(param['phi']) * y[-1] + param['e']
            y1 = param['r'] * np.sin(param['theta']) * x[-1] + param['s'] * np.cos(param['phi']) * y[-1] + param['f']
            x.append(x1)
            y.append(y1)

    return x, y

def display(iterations: int, params: list, title: str = 'Fractal Visualization', color_index: int = 1) -> None:
    
    x, y = generate_fractal(iterations, params)
    colors = plt.cm.viridis(np.linspace(0, 1, 14))  
    chosen_color = colors[color_index % 14]  

    plt.figure()
    plt.scatter(x, y, color=chosen_color, s=0.1)  
    plt.title(title)
    plt.axis('off')  
    plt.show()
```
