```python
import numpy as np
import matplotlib.pyplot as plt

# Define the transformation parameters for several fractals

# Parameters for "Maple Leaf" (1.4.1)
maple_leaf_params = [
    (0.140, 0.010, 0.000, 0.510, -0.080, -1.310),
    (0.430, 0.520, -0.450, 0.500, 1.490, -0.750),
    (0.450, -0.490, 0.470, 0.470, -1.620, -0.740),
    (0.490, 0.000, 0.000, 0.510, 0.020, 1.620)
]

# Parameters for "Spiral" (1.4.2)
spiral_params = [
    (0.787879, -0.424242, 0.242424, 0.859848, 1.758647, 1.408065),
    (-0.121212, 0.257576, 0.151515, 0.053030, -6.721654, 1.377236),
    (0.181818, -0.136364, 0.090909, 0.181818, 6.086107, 1.568035)
]

# Parameters for "Mandelbrot-like" (1.4.3)
mandelbrot_like_params = [
    (0.2020, -0.8050, -0.6890, -0.3420, -0.3730, -0.6530),
    (0.1380, 0.6650, -0.5020, -0.2220, 0.6600, -0.2770)
]

# Parameters for "Fractal Tree" (1.4.4)
fractal_tree_params = [
    (0.0500, 0.0000, 0.0000, 0.4000, -0.0600, -0.4700),
    (-0.0500, 0.0000, 0.0000, -0.4000, -0.0600, -0.4700),
    (0.0300, -0.1400, 0.0000, 0.2600, -0.1600, -0.0100),
    (-0.0300, 0.1400, 0.0000, -0.2600, -0.1600, -0.0100),
    (0.5600, 0.4400, -0.3700, 0.5100, 0.3000, 0.1500),
    (0.1900, 0.0700, -0.1000, 0.1500, -0.2000, 0.2800),
    (-0.3300, -0.3400, -0.3300, 0.3400, -0.5400, 0.3900)
]

# Helper function to generate fractals
def generate_fractal(params, num_points=100000):
    x, y = 0, 0
    xs, ys = [], []
    
    for _ in range(num_points):
        idx = np.random.randint(len(params))  # Generate a random index
        p = params[idx]  # Access the tuple using the index
        x_new = p[0] * x + p[1] * y + p[4]
        y_new = p[2] * x + p[3] * y + p[5]
        x, y = x_new, y_new
        xs.append(x)
        ys.append(y)
    
    return xs, ys

# Generate each fractal
maple_leaf_xs, maple_leaf_ys = generate_fractal(maple_leaf_params)
spiral_xs, spiral_ys = generate_fractal(spiral_params)
mandelbrot_like_xs, mandelbrot_like_ys = generate_fractal(mandelbrot_like_params)
fractal_tree_xs, fractal_tree_ys = generate_fractal(fractal_tree_params)

# Plotting all fractals
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Maple Leaf
axs[0, 0].scatter(maple_leaf_xs, maple_leaf_ys, color='green', s=0.1)
axs[0, 0].set_title("Maple Leaf")
axs[0, 0].axis('off')

# Spiral
axs[0, 1].scatter(spiral_xs, spiral_ys, color='blue', s=0.1)
axs[0, 1].set_title("Spiral")
axs[0, 1].axis('off')

# Mandelbrot-like
axs[1, 0].scatter(mandelbrot_like_xs, mandelbrot_like_ys, color='red', s=0.1)
axs[1, 0].set_title("Mandelbrot-like")
axs[1, 0].axis('off')

# Fractal Tree
axs[1, 1].scatter(fractal_tree_xs, fractal_tree_ys, color='brown', s=0.1)
axs[1, 1].set_title("Fractal Tree")
axs[1, 1].axis('off')

# Display the plots
plt.tight_layout()
plt.show()
```
