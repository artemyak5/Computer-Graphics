<p>4. У розроблених програмах показати фінальний результат побудови кожного з наведених фракталів.</p>

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the transformation parameters for several fractals
maple_leaf_params = [
    (0.140, 0.010, 0.000, 0.510, -0.080, -1.310),
    (0.430, 0.520, -0.450, 0.500, 1.490, -0.750),
    (0.450, -0.490, 0.470, 0.470, -1.620, -0.740),
    (0.490, 0.000, 0.000, 0.510, 0.020, 1.620)
]

spiral_params = [
    (0.787879, -0.424242, 0.242424, 0.859848, 1.758647, 1.408065),
    (-0.121212, 0.257576, 0.151515, 0.053030, -6.721654, 1.377236),
    (0.181818, -0.136364, 0.090909, 0.181818, 6.086107, 1.568035)
]

mandelbrot_like_params = [
    (0.2020, -0.8050, -0.6890, -0.3420, -0.3730, -0.6530),
    (0.1380, 0.6650, -0.5020, -0.2220, 0.6600, -0.2770)
]

fractal_tree_params = [
    (0.0500, 0.0000, 0.0000, 0.4000, -0.0600, -0.4700),
    (-0.0500, 0.0000, 0.0000, -0.4000, -0.0600, -0.4700),
    (0.0300, -0.1400, 0.0000, 0.2600, -0.1600, -0.0100),
    (-0.0300, 0.1400, 0.0000, -0.2600, -0.1600, -0.0100),
    (0.5600, 0.4400, -0.3700, 0.5100, 0.3000, 0.1500),
    (0.1900, 0.0700, -0.1000, 0.1500, -0.2000, 0.2800),
    (-0.3300, -0.3400, -0.3300, 0.3400, -0.5400, 0.3900)
]
second_variant_tree_params = [
    (0.0500, 0.6000, 0.0500, 0.5000, 0.0000, 1.0000),
    (0.0500, 0.5000, -0.5000, 0.4500, 0.0000, 0.6000),
    (0.6000, 0.5000, 0.5000, 0.5500, 0.0000, 1.1000),
    (0.5000, 0.5500, 0.4000, 0.4000, 0.0000, 1.0000),
    (0.5500, 0.4000, 0.4000, 0.4000, 0.0000, 0.7000)
]

third_variant_tree_params = [
    (0.0100, 0.0000, 0.0000, 0.4500, 0.0000, 0.0000),
    (-0.0100, 0.0000, 0.0000, -0.4500, 0.0000, 0.4000),
    (0.4200, -0.4200, 0.4200, 0.4200, 0.0000, 0.4000),
    (0.4200, 0.4200, -0.4200, 0.4200, 0.0000, 0.4000)
]

fourth_variant_tree_params = [
    (0.1950, -0.4880, 0.3440, 0.4430, 0.4431, 0.2452),
    (0.4620, 0.4140, -0.2520, 0.3610, 0.2511, 0.5692),
    (-0.6370, 0.0000, 0.0000, 0.5010, 0.8562, 0.2512),
    (-0.0350, 0.0700, -0.4690, 0.0220, 0.4884, 0.5069),
    (-0.0580, -0.0700, 0.4530, -0.1110, 0.5976, 0.0969)
]

fractal_leaf_params = [
    (0.0000, 0.2439, 0.0000, 0.3053, 0.0000, 0.0000),
    (0.7248, 0.0337, -0.0253, 0.7426, 0.2060, 0.2538),
    (0.1583, -0.1297, 0.3550, 0.3676, 0.1383, 0.1750),
    (0.3386, 0.3694, 0.2227, -0.0756, 0.0679, 0.0826)
]
sand_dollar_snowflake_params = [
    (0.38200, 0.00000, 0.00000, 0.38200, 0.30900, 0.57000),
    (0.11800, -0.36330, 0.36330, 0.11800, 0.36330, 0.33060),
    (0.11800, 0.36330, -0.36330, 0.11800, 0.51870, 0.69400),
    (-0.30900, -0.22450, 0.22450, -0.30900, 0.60700, 0.30900),
    (-0.30900, 0.22450, -0.22450, -0.30900, 0.70160, 0.53350),
    (0.38200, 0.00000, 0.00000, -0.38200, 0.30900, 0.67700)
]
fern_params = [
    (0.0, 0.0, 0.0, 0.16, 0.0, 0.0),
    (0.2, -0.26, 0.23, 0.22, 0.0, 1.6),
    (-0.15, 0.28, 0.26, 0.24, 0.0, 0.44),
    (0.75, 0.04, -0.04, 0.85, 0.0, 1.6)
]

def generate_fractal(params, num_points=100000):
    x, y = 0, 0
    xs, ys = [], []
    for _ in range(num_points):
        idx = np.random.randint(len(params))
        p = params[idx]
        x_new = p[0] * x + p[1] * y + p[4]
        y_new = p[2] * x + p[3] * y + p[5]
        x, y = x_new, y_new
        xs.append(x)
        ys.append(y)
    return xs, ys

# Setup the plot for multiple fractals
fig, axs = plt.subplots(3, 4, figsize=(20, 15))  # Adjusted to 3x4 grid layout
axs = axs.flatten()  # Ensure axs is a 1D array for easy indexing

# List of colors for each fractal
colors = [
    'red', 'green', 'blue', 'yellow', 'purple',
    'orange', 'pink', 'brown', 'grey', 'cyan'
]

fractals = [
    (maple_leaf_params, 'Maple Leaf', axs[0]),
    (spiral_params, 'Spiral', axs[1]),
    (mandelbrot_like_params, 'Mandelbrot-like', axs[2]),
    (fractal_tree_params, 'Fractal Tree', axs[3]),
    (second_variant_tree_params, 'Second Variant Tree', axs[4]),
    (third_variant_tree_params, 'Third Variant Tree', axs[5]),
    (fourth_variant_tree_params, 'Fourth Variant Tree', axs[6]),
    (fractal_leaf_params, 'Fractal Leaf', axs[7]),
    (sand_dollar_snowflake_params, 'Sand Dollar-Snowflake', axs[8]),
    (fern_params, 'Fern', axs[9])
]

for i, (params, title, ax) in enumerate(fractals):
    xs, ys = generate_fractal(params)
    ax.scatter(xs, ys, color=colors[i], s=0.1)  # Use a unique color for each fractal
    ax.set_title(title)
    ax.axis('off')

# Optionally hide empty subplot axes
for empty_ax in axs[10:]:
    empty_ax.axis('off')

plt.tight_layout()
plt.show()
```

Для деяких фракталів потрібно буде виводити не кожну точку ітерації, а кілька точок згенерованих протягом M (задає розробник програми) ітерацій, оскільки для досить великої кількості точок виведення по одній точці буде тривалим у часі.</p>
